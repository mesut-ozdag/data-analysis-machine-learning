## PLOTTING WITH SEABORN
matplotlib can be a fairly low-level tool. You assemble a plot from its base components: the data display (i.e., the type of plot: line, bar, box, scatter, contour, etc.), legend, title, tick labels, and other annotations.

In pandas, we may have multiple columns of data, along with row and column labels. pandas itself has built-in methods that simplify creating visualizations from Data! Frame and Series objects. Another library is seaborn, a high-level statistical graphics library built on matplotlib. seaborn simplifies creating many common visualization types.

Seaborn, a Python data visualization library, offers a range of built-in datasets that are perfect for practicing and demonstrating various data science concepts. These datasets are designed to be simple, intuitive, and easy to work with, making them ideal for beginners and experienced data scientists alike.


To read a CSV file with Seaborn, you need to use the pandas.read_csv() function to load the data into a Pandas DataFrame, as Seaborn primarily works with Pandas data structures; essentially, you import pandas, read the CSV using pd.read_csv(), and then use the resulting DataFrame with Seaborn plotting functions.

data=data: specifies the DataFrame to use for plotting. 

x="column_name_x": selects the column to plot on the x-axis. 

y="column_name_y": selects the column to plot on the y-axis. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("iris.csv") 

# Create a scatterplot using Seaborn
sns.scatterplot(x="PetalLength", y="PetalWidth", data=df, hue="Name") 

# Show the plot
plt.show() 
```

# Distribution plots

This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots. The kind parameter selects the approach to use:


histplot
Plot a histogram of binned counts with optional normalization or smoothing.

kdeplot
Plot univariate or bivariate distributions using kernel density estimation.

rugplot
Plot a tick at each observation value along the x and/or y axes.

ecdfplot
Plot empirical cumulative distribution functions.

jointplot
Draw a bivariate plot with univariate marginal distributions.

```python
# See the API documentation for the axes-level functions for more details about
# the breadth of options available for each plot kind.
# The default plot kind is a histogram:
# try with histplot, ecdfplot.
sns.displot([tips['total_bill']])
```

```python
# There is another plot kind; in addition to histograms and kernel density estimates (KDEs),
# While in histogram mode, it is also possible to add a KDE curve:
# A univariate distribution plot or histogram is a graph that shows the distribution of
# a single variable
sns.displot([tips['total_bill']], kde=True) # Histogram
```

```python
# To draw a bivariate plot, assign both x and y:
sns.displot(x='total_bill', y='tip', data=tips) # Bivariate histogram/distribution plot
```

```python
# Draw a plot of two variables with bivariate and univariate graphs.
sns.jointplot(x='total_bill', y='tip', data=tips) # Bivariate histogram/distribution plot
```

```python
# `kind` must be one of ['hist', 'kde', 'ecdf']
# try hist and kde
sns.displot(x='total_bill', y='tip', data=tips, kind='hist') # Bivariate histogram/distribution plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # Bivariate histogram/distribution plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # Bivariate histogram/distribution plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') # Bivariate histogram/distribution plot
```

```python
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location
# of individual observations in an unobtrusive way.
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde=False)
```

# Matrix plots (heatmap)
## HEATMAP
Plot rectangular data as a color-encoded matrix.

This is an Axes-level function and will draw the heatmap into the currently-active Axes.

A Seaborn heatmap is a data visualization technique that uses color to represent the magnitude of values in a matrix or a two-dimensional dataset. It's a powerful way to visualize patterns, correlations, and relationships between variables in a visually appealing and easy-to-understand manner.

Here's a breakdown of what a Seaborn heatmap is and how it works:

Purpose: Heatmaps display the intensity or magnitude of a phenomenon as color, making them ideal for visualizing large datasets and complex relationships.

Data Representation: The heatmap takes a matrix as input, where each cell represents the intersection of two variables. The color intensity of each cell corresponds to the value of the data point.

Color Mapping: A colormap (a range of colors) is used to map the data values to colors. Typically, higher values are represented by warmer colors (e.g., red), while lower values are represented by cooler colors (e.g., blue).

Correlation Analysis: One of the most common uses of heatmaps is to visualize correlation matrices, which show the relationships between multiple variables. Positive correlations are typically represented by warm colors, while negative correlations are represented by cool colors. 

Visual Interpretation: By looking at the heatmap, you can quickly identify patterns, trends, and outliers in the data. Areas of high intensity stand out, making it easy to spot areas of interest.

sns.heatmap(): This function creates the heatmap.

df: The input DataFrame containing the data to be visualized.

annot=True: Annotates the cells with their corresponding values.

cmap='coolwarm': Specifies the colormap to use.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Create the heatmap
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.show()
```

# Regression plots

Plot data and regression model fits across a FacetGrid.

This function combines regplot() and FacetGrid. It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.

When thinking about how to assign variables to different facets, a general rule is that it makes sense to use hue for the most important comparison, followed by col and row. However, always think about your particular dataset and the goals of the visualization you are creating.

There are a number of mutually exclusive options for estimating the regression model. See the tutorial for more information.

The parameters to this function span most of the options in FacetGrid, although there may be occasional cases where you will want to use that class and regplot() directly.

Seaborn's lmplot is a powerful function for visualizing linear relationships between variables. Here's a breakdown of how it works:

What it does:

Scatterplot with Regression Line:
lmplot creates a scatterplot of two variables, and then fits a linear regression model to the data. The resulting regression line is plotted on the scatterplot, along with a 95% confidence interval band.

Regression plots in seaborn can be easily implemented with the help of the lmplot() function. lmplot() can be understood as a function that basically creates a linear model plot. lmplot() makes a very simple linear regression plot.It creates a scatter plot with a linear fit on top of it.

FacetGrid:
lmplot is built on top of Seaborn's FacetGrid, which allows you to easily create multiple subplots based on categorical variables. This means you can visualize how the relationship between your variables changes across different categories.

Key arguments:

x, y: The names of the variables to plot on the x and y axes.

data: The DataFrame containing the data.

hue: The name of a categorical variable to use for color-coding the points and regression lines.

col, row: The names of categorical variables to use for creating subplots in columns and rows.

fit_reg: A boolean value indicating whether to fit a regression line (default is True).

ci: The confidence interval to plot around the regression line (default is 95%).

scatter_kws, line_kws: Dictionaries of keyword arguments to pass to plt.scatter and plt.plot, respectively, for customizing the appearance of the plot.

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'],
          scatter_kws={'s':75})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex',
          aspect=0.6, height=6)
```
