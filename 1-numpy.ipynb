# Numerical Computing with NumPy
Although the Python interpreter itself already brings a rich variety of data structures
with it, NumPy and other libraries add to these in a valuable fashion. This chapter
focuses on NumPy, which provides a multidimensional array object to store homogeneous or heterogeneous data arrays and supports vectorization of code.

NumPy, short for Numerical Python, serves as the cornerstone for various scientific and numerical computations in the Python programming language. Given that machine learning revolves around transforming data into numerical representations and subsequently deciphering patterns, NumPy frequently plays a significant role in this context.

In Python, both arrays and NumPy arrays are used to store collections of data, but they have key differences:
Data Type:

Arrays:
Python's built-in array module creates arrays that store elements of a single data type. You specify the type when creating the array, and it can't be changed. This makes them efficient for storing homogeneous data.
NumPy Arrays:
NumPy arrays (ndarray) can also store elements of a single data type, but they are more flexible. NumPy automatically handles type casting when possible, allowing you to mix data types in some cases.

Functionality:

Arrays:
The array module offers basic operations like appending, extending, and removing elements. However, it lacks advanced mathematical and scientific operations.

NumPy Arrays:
NumPy provides a vast array of functions for mathematical operations, linear algebra, Fourier transforms, random number generation, and much more. This makes them ideal for scientific computing and data analysis.

Performance:

Arrays:
Python arrays are not optimized for performance, especially for large datasets.

NumPy Arrays:
NumPy arrays are implemented in C, making them significantly faster than Python arrays for most operations. They also use less memory, which is crucial for handling large datasets.

Broadcasting:

Arrays: Python arrays do not support broadcasting.

NumPy Arrays: NumPy arrays support broadcasting, which allows you to perform operations on arrays with different shapes as long as certain rules are met.

If you are curious how NumPy achieves speed, it is by using process called Vectorization. Vectorization allows operations to be applied on whole arrays/data structures instead of individual elements. Which will avoid loops as loops will cause bottlenecks. NumPy achieves vectorization through a process called broadcasting.

Overall, NumPy arrays are the preferred choice for most data science and numerical computing tasks due to their superior performance, functionality, and flexibility.
Here's an example to illustrate the difference in performance:


```python
import array
import numpy as np
import time

# Create a Python array and a NumPy array
python_array = array.array('i', range(1000000))
numpy_array = np.arange(1000000)

# Time the sum operation for Python array
start_time = time.time()
sum(python_array)
end_time = time.time()
print("Python array sum time:", end_time - start_time)

# Time the sum operation for NumPy array
start_time = time.time()
np.sum(numpy_array)
end_time = time.time()
print("NumPy array sum time:", end_time - start_time)
```

## Regular NumPy Arrays
Composing array structures with list objects works, somewhat. But it is not really convenient, and the list class has not been built with this specific goal in mind. It has rather a much broader and more general scope. The array class is a bit more specialized, providing some useful features for working with arrays of data. However, a truly specialized class could be really beneficial to handle array-type structures.

### The Basics
numpy.ndarray is just such a class, built with the specific goal of handling ndimensional arrays both conveniently and efficiently—i.e., in a highly performant manner. The basic handling of instances of this class is again best illustrated by examples:

```python
import numpy as np  
```

```python
# Creates an ndarray object out of a list object with floats.
a = np.array([0, 0.5, 1.0, 1.5, 2.0])  
a
```

```python
type(a)  
```

```python
# Creates an ndarray object out of a list object with strs.
a = np.array(['a', 'b', 'c'])  
a
```

```python
# np.arange() works similar to range()
a = np.arange(2, 20, 2)  
a
```

```python
# … but takes as additional input the dtype parameter.
a = np.arange(8, dtype=float)  
a
```

```python
# With one-dimensional ndarray objects, indexing works as usual.
a[5:]  
```

```python
a[:2]  
```

```python
# A major feature of the ndarray class is the multitude of built-in methods. For instance:
# The sum of all elements
a.sum()  
```

```python
# The standard deviation of the elements
a.std()  
```

```python
# The cumulative sum of all elements (starting at index position 0).
a.cumsum()  
```

```python
# Another major feature is the (vectorized) mathematical operations defined on ndarray objects:
# Scalar multiplication with list objects leads to a repetition of elements.
l = [0., 0.5, 1.5, 3., 5.]
2 * l  
```

```python
# This interprets the elements of the ndarray as the powers.
2 ** a  
```

### Note:
There are a number of ways to initialize (instantiate) ndarray objects. One is as presented before, via np.array. However, this assumes that all elements of the array are already available. In contrast, one might like to have the ndarray objects instantiated first to populate them later with results generated during the execution of code. To this end, one can use the following functions:

the order parameter: Whether to store multi-dimensional data in row-wise (C-style) or column-wise (Fortran-style) order in memory.

```python
# Creates an ndarray object prepopulated with zeros.
c = np.zeros((2, 3), dtype='i', order='C')  
c
```

```python
# Creates an ndarray object prepopulated with ones.
c = np.ones((2, 3, 4), dtype='i', order='C')  
c
```

```python
# Creates a square matrix as an ndarray object with the diagonal populated by ones.
np.eye(5)  
```

```python
# Creates a one-dimensional ndarray object with evenly spaced intervals between numbers;
# parameters used are start, end, and num (number of elements).
g = np.linspace(5, 15, 12) 
g
```

### Note
Here, it becomes obvious how NumPy specializes the construction of arrays with the ndarray class, in comparison to the list -based approach:

• The ndarray object has built-in dimensions (axes).

• The ndarray object is immutable; its length (size) is fixed.

• It only allows for a single data type (np.dtype) for the whole array.

The array class by contrast shares only the characteristic of allowing for a single data
type (type code, dtype).

### Reshaping, Resizing, Stacking, Flattening
Although ndarray objects are immutable by default, there are multiple options to reshape and resize such an object. While reshaping in general just provides another view on the same data, resizing in general creates a new (temporary) object. First, some examples of reshaping:

```python
g = np.arange(15)
```

```python
g
```

```python
g.shape  
```

```python
np.shape(g) 
```

```python
# Reshaping to two dimensions (memory view).
g.reshape((3, 5))  
```

```python
# Creating a new object.
h = g.reshape((5, 3))  
h
```

```python
# The transpose of the new ndarray object.
h.T  
```

```python
h.transpose()  
```

### Note
During a reshaping operation, the total number of elements in the ndarray object is unchanged. During a resizing operation, this number changes—it either decreases (“down-sizing”) or increases (“up-sizing”). Here some examples of resizing:

```python
g
```

```python
# Two dimensions, down-sizing.
np.resize(g, (3, 1))  
```

```python
np.resize(g, (1, 5))  
```

```python
np.resize(g, (2, 5))  
```

```python
# Two dimensions, up-sizing.
n = np.resize(g, (5, 4))  
n
```

```python
# Stacking is a special operation that allows the horizontal or vertical combination of
# two ndarray objects. However, the size of the “connecting” dimension must be the same:
h
```

```python
# Horizontal stacking of two ndarray objects.
np.hstack((h, 2 * h))  
```

```python
# Vertical stacking of two ndarray objects.
np.vstack((h, 0.5 * h))  
```

```python
# Another special operation is the flattening of a multidimensional ndarray object to a
# one-dimensional one. One can choose whether the flattening happens row-by-row (C order)
# or column-by-column (F order):
h
```

```python
# The default order for flattening is C.
h.flatten()  
```

```python
h.flatten(order='C')  
```

```python
# Flattening with F order.
h.flatten(order='F')  
```

```python
# The flat attribute provides a flat iterator (C order).
for i in h.flat:  
    print(i, end=',')
```

### Boolean Arrays
Comparison and logical operations in general work on ndarray objects the same way, element-wise, as on standard Python data types. Evaluating conditions yield by default a Boolean ndarray object (dtype is bool):

```python
h
```

```python
h > 8  
```

```python
h <= 7  
```

```python
h == 5  
```

```python
(h == 5).astype(int)  
```

```python
(h > 4) & (h <= 12)  
```

```python
# Such Boolean arrays can be used for indexing and data selection. Notice that the
# following operations flatten the data:
# Give me all values greater than 8:
h[h > 8]  
```

```python
h[(h > 4) & (h <= 12)]  
```

```python
h[(h < 4) | (h >= 12)]  
```

```python
# A powerful tool in this regard is the np.where() function, which allows the definition
# of actions/operations depending on whether a condition is True or False. The result
# of applying np.where() is a new ndarray object of the same shape as the original one:
# In the new object, set 1 if True and 0 otherwise:
np.where(h > 7, 1, 0)  
```

```python
# In the new object, set even if True and odd otherwise.
np.where(h % 2 == 0, 'even', 'odd')  
```

```python
# In the new object, set two times the h element if True and half the h element otherwise.
np.where(h <= 7, h * 2, h / 2)  
```

### Structured NumPy Arrays
The specialization of the ndarray class obviously brings a number of valuable benefits with it. However, a too narrow specialization might turn out to be too large a burden to carry for the majority of array-based algorithms and applications. Therefore, NumPy provides structured ndarray and record recarray objects that allow you to have a different dtype per column. What does “per column” mean? Consider the following initialization of a structured ndarray object:

```python
# The complex dtype is composed.
dt = np.dtype([('Name', 'S10'), ('Age', 'i4'),
               ('Height', 'f'), ('Children/Pets', 'i4', 2)])  
```

```python
dt  
```

```python
# An alternative syntax to achieve the same result.
dt = np.dtype({'names': ['Name', 'Age', 'Height', 'Children/Pets'],
             'formats':'O int float int,int'.split()})  
```

```python
dt  
```

```python
# The structured ndarray is instantiated with two records.
s = np.array([('Smith', 45, 1.83, (0, 1)),
              ('Jones', 53, 1.72, (2, 2))], dtype=dt)  
```

```python
s  
```

```python
type(s)  
```

### Note:
In a sense, this construction comes quite close to the operation for initializing tables in a SQL database: one has column names and column data types, with maybe some additional information (e.g., maximum number of characters per str object). The single columns can now be easily accessed by their names and the rows by their index values:

```python
s['Name']  
```

```python
s['Height'].mean()  
```

```python
s[0]  
```

```python
s[1]['Age']  
```

### Note:
In summary, structured arrays are a generalization of the regular ndarray object type in that the data type only has to be the same per column, like in tables in SQL databases. One advantage of structured arrays is that a single element of a column can be another multidimensional object and does not have to conform to the basic NumPy data types.

## Vectorization of Code
Vectorization is a strategy to get more compact code that is possibly executed faster. The fundamental idea is to conduct an operation on or to apply a function to a complex object “at once” and not by looping over the single elements of the object. In Python, functional programming tools such as map() and filter() provide some basic means for vectorization. However, NumPy has vectorization built in deep down in its core.

As demonstrated in the previous section, simple mathematical operations—such as calculating the sum of all elements—can be implemented on ndarray objects directly (via methods or universal functions). More general vectorized operations are also possible. For example, one can add two NumPy arrays element-wise as follows:

```python
np.random.seed(100)
r = np.arange(12).reshape((4, 3))  
s = np.arange(12).reshape((4, 3)) * 0.5  
```
